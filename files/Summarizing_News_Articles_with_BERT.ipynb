#Extract text from news link with newspaper3k

```python
from newspaper import fulltext
import requests
url = "https://www.ndtv.com/entertainment/10-movies-on-ott-platforms-to-binge-watch-on-independence-day-2509669#pfrom=home-ndtv_icymi"
text = fulltext(requests.get(url).text)
print(text)
```
